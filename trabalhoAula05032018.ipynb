```python
import numpy as np
from matplotlib import pyplot as plt
import operator
def media(x, y):
    return (x + y) /2

data = np.genfromtxt('C:\\Users\\Zatara\\anaconda\\iris.data.txt', delimiter=',', usecols=(0,1,2,3))

data_r = [reduce(media, data[i]) for i,_ in enumerate(data)]
print data_r
print data

primeiros50 = data_r[:50]
ultimos50 = data_r[50:100]
quantidade = [i for i, _ in enumerate(primeiros50)]
plt.plot(quantidade, primeiros50, color='green', marker='>', linestyle='--', label="Comp. Sepala Setosa")
plt.plot(quantidade, ultimos50, color='blue', marker='o', linestyle='--', label="Comp. Sepala Versicolor")
plt.title("Compara Flor")
# add a label to the y-axis
plt.ylabel("Tamanho Flor")
##plt.legend(loc=0)
plt.show()

```
