Probabilidade
É o ramo da matematica de ocorrencia de experimentos.
São calculadas de certo modo, que podemos afirmar que se trata da quantificacao da incerteza.
É por meio da probabilidade que podemos saber desde as chances de sair cara ou coroa em um lançamento de uma moeda
até os erros em uma pesquisa.
Uma probabilidade é representada por um valor real entre 0 e 1 que é uma forma de representar
quantitativamente a noção qualitativa de que algumas coisas são mais prováveis que outras.
As "coisas" para as quais atribuimos probabilidade são chamadas de eventos. Assim , se 'E' represente um evento,
'P(E)' representa a probabilidade desse evento acontecer.
A situação onde 'E' pode ou não ocorrer é chamada de TESTE.
Por exemplo:
    Suponha que voce tenha um dado e queira saber a probabilidade de sair um 6 em um lançamento. Cada lan,amento é um TESTE.
    Cada vez que o 6 aparece é um SUCESSO e cada vez que aparece outro número, é uma FALHA.
   
A seguir, alguns conceitos e termos:
    
    
    EXPERIMENTO ALEATORIO
    
        Qualquer experiencia ou teste cujo resultado não seja conhecido. Por exemplo:
        Jogar uma moeda e observar a face.
        

    PONTO AMOSTRAL
        
        é qualquer resultado possivel em um experimento aleatório. Por exemplo:
        A face da moeda ser cara ou coroa.
        
    ESPAÇO AMOSTRAL
        
        É o conjunto de todos os pontos amostrais de um experimento aleatório. É representado pela letra omega.
        Exemplo:
        lançamento da moeda -> Ω = {cara, coroa}
        

CALCULO DE PROBABILIDADE
    As probabilidades são calculadas dividindo-se o número de resultados favoraveis pelo número de resultados possíveis.
    
    P = N(E) / n(Ω)
    
    Sendo 'E' um evento e Ω espaço amostral.
    Por exemplo:
        No lançcamento de um dado, qual a probabilidade de sair o numero 1?
        E = sair número 1 => N(E) = 1
        Ω = {range(1,7)} => n(Ω) = 6
        Assim:
        
        P(E) = 1 / 6
        P(E) = 0,16666
        P(E) = 16,6%
        

Probabilidade da união de dois eventos
    Dado dois eventos A e B de um espaço amostral S, a probabilidade de ocorrer A ou B é dada por:
        
        P(A U B) = P(A) + P(B) - P(A ∩ B)
        
    Exemplo:
        Em um saco existem 10 bolas numeradas de 1 a 10.Retirando-se uma bola ao acaso, qual a probabilidade de ocorrer multiplos de 2 ou de 3?
        
        P(A U B) = P(A) + P(B) - P(A ∩ B)
        
        P(A) = 5 / 10
        
        P(B) = 3 / 10
        
        P(A ∩ B) = 1 / 10
        
        P(A U B) = 7 / 10
        

Probabilidade de um evento complementar
    Considere o seguinte exemplo:
        O espaço amostral de um lançamento de dado é igual a 6.
        Se considerarmos o evento A como valores menores que 5 teremos n(A) = 4 composto por {1,2,3,4}.
        Nessa situação, o evento complementar B possui n(B) = 2, composto por {5,6}.
        A união dos eventos A e B formam o espaço amostral S e a intersecção dos dois eventos (A ∩ B) forma um conjunto vazio.
        Exemplo:
        
        No lançamento simultâneo de 2 dados, vamos determinar a probabilidade de nao sair soma 4, evento A.
            
            n(S) = 36
            
            A = {(1,3), (3,1), (2,2)}
            
            n(A) = 3
        

DEPENDENCIA E INTERDEPENDENCIA DE EVENTOS
    A grosso modo, podemos dizer que dois eventos são dependentes se saber algo sobre a ocorrência de um me diz algo sobre a ocorrência de outro.
    De outro modo, eles são interdependentes.
    Por exemplo:
        Se eu jogar uma moeda 2 vezes, saber que saiu cara na primeira vez, não me diz nada sobre a possibilidade de sair cara na segunda vez. Estes 2 eventos são interdependentes.Por outro lado, saber que saiu cara na primeira vez me informa sobre a possibilidade de sair cara em ambas jogadas. Neste caso os eventos são dependentes.
        
        Matematicamente, dizemos que dois eventos E e F são interdependentes se a possibilidade de ambos ocorrerem for o produto das probabilidades de cada um ocorrer.
        
        P(E,F) = P(E) * P(F)
        
        No exemplo da moeda, a probabilidade de sair cara é 1/2 e a probabilidade de sair 2 coroas é 1/4 , mas a probabilidade de sair cara na primeira e coroa nas 2 vezes é zero.
        
        

PROBABILIDADE CONDICIONAL
    Quando 2 eventos são independentes:
        P(E,F) = P(E) * P(F)
        
    Se forem dependentes temos:
        P(E|F) = P(E,F) / P(F)
        
        ou
        
        P(E,F) = P(E|F) * P(F)
