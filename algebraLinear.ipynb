```python
def vector_add(v, w):
    ##
    """adds corresponding elements"""
    return [v_i + w_i
            for v_i, w_i in zip(v, w)]

print vector_add([100, 9, 69, 48], [38, 86, 36, 82], [85, 23, 6, 16])

def vector_subtract(v, w):
    ##
    """subtracts corresponding elements"""
    return [v_i - w_i
            for v_i, w_i in zip(v, w)]

print vector_subtract([2,6,5],[4,8,2])
```

```python
def vector_sum(vectors):
    """sums all corresponding elements"""
    result = vectors[0] # start with the first vector
    for vector in vectors[1:]: # then loop over the others
        result = vector_add(result, vector) # and add them to the result
    return result

print vector_sum([[2,5,7],[4,8,9]])
```

```python
def scalar_multiply(c, v):
    """c is a number, v is a vector"""
    return [c * v_i for v_i in v]

print scalar_multiply(3, [3,2,1])
```

```python
def vector_mean(vectors):
    """compute the vector whose ith element is the mean of the
                ith elements of the input vectors"""
    n = len(vectors)
    return scalar_multiply(1/n, vector_sum2(vectors))

print vector_mean([[2,5,7],[4,8,9],[2,6,7]])
```

```python
def dot(v, w):
    """v_1 * w_1 + ... + v_n * w_n"""
    return sum(v_i * w_i for v_i, w_i in zip(v, w))

print dot([5,4,9], [3,8,5])
```

```python
def sum_of_squares(v):
    """v_1 * v_1 + ... + v_n * v_n"""
    return dot(v, v)

def magnitude(v):
    return math.sqrt(sum_of_squares(v))

print magnitude([1,2,3])
```

```python
def squared_distance(v, w):
    """(v_1 - w_1) ** 2 + ... + (v_n - w_n) ** 2"""
    return sum_of_squares(vector_subtract(v, w))


def distance(v, w):
    return math.sqrt(squared_distance(v, w))

print distance([1,2,3],[3,2,1])
```

```python
A = [
        [1,2,3],
        [4,5,6]
]

B = [
    [1,2],
    [3,4],
    [5,6]
]

def shape(A):
    num_rows = len(A)
    num_cols = len(A[0]) if A else 0 # number of elements in first row
    return num_rows, num_cols

print shape(A)
print shape(B)
```

```python
def get_row(A, i):
    return A[i] # A[i] is already the ith row


def get_column(A, j):
    return [A_i[j] # jth element of row A_i
            for A_i in A] # for each row A_i

print get_row(A, 1)
print get_column(B,1)
```

```python
def make_matrix(num_rows, num_cols, entry_fn):
    """returns a num_rows x num_cols matrix
            whose (i,j)th entry is entry_fn(i, j)"""
    return [[entry_fn(i, j) for j in range(num_cols)] for i in range(num_rows)] # create one list for each i

def is_diagonal(i, j):
    """1's on the 'diagonal', 0's everywhere else"""
    return 1 if i == j else 0

identity_matrix = make_matrix(5, 5, is_diagonal)

print identity_matrix
```
