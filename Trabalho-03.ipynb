```python
import numpy as np
from random import randint as rand

def make_matrix(x, y,mn,mx):
    return [[rand(mn,mx) for j in range(y)] for i in range(x)] 

def get_row(A, i):
    return A[i] 
def get_column(A, j):
    return [A_i[j] for A_i in A]

def shape(A):
    num_rows = len(A)
    num_cols = len(A[0]) if A else 0
    return num_rows, num_cols

def vet_med(A, eixo):
    lin, col = shape(A)
    if eixo=="linha":
        return[np.mean(A[i]) for i in range(0, lin)]
    elif eixo=="coluna":
        return [np.mean(get_column(A, i)) for i in range(0,col)]
    else:
        print "Informe 'linha' ou 'coluna'!"

matriz = make_matrix(3,4,0,100)
print matriz
print vet_med(matriz,'linha')     
```

```python
import math

data = np.genfromtxt('C:\\apagar\\iris.data.txt', delimiter=',', usecols=(0,1,2,3))

def calc_med(v):
    return sum(v)/len(v)

def vector_subtract(v, w):
    return [v_i - w_i for v_i, w_i in zip(v, w)]

def dot(v, w):
    return sum(v_i * w_i for v_i, w_i in zip(v, w))

def sum_of_squares(v):
    return dot(v, v)

def magnitude(v):
    return math.sqrt(sum_of_squares(v))

def distance(v,w):
    return magnitude(vector_subtract(v,w))

setosa=[np.mean(get_column(data[0:50], i)) for i in range(0,4)]
versicolor=[np.mean(get_column(data[50:100], i)) for i in range(0,4)]
virginica=[np.mean(get_column(data[100:150], i)) for i in range(0,4)]

print "\nsetosa: ",setosa,"\nversicolor", versicolor, "\nvirginica", virginica, "\n"

nova=[[5.0,3.0,1.0,0.3],
      [6.0,3.0,4.0,1.0],
      [6.0,3.0,5.0,2.0]]

dist_set=distance(setosa, nova[0])
dist_ver=distance(versicolor, nova[0])
dist_vir=distance(virginica, nova[0])

print "\nDistancias:\n",dist_set,"\n",dist_ver,"\n",dist_vir,"\n"

tipo="Setosa"
if dist_ver < dist_set:
    tipo="Versicolor"
if dist_vir < dist_ver:
    tipo="Virginica"
print "Nova: ",tipo
```
